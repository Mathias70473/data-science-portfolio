```python
## Library importation

import pandas as pd
```

```python
## Df creation for each quarter
df_2019_03=pd.read_excel(r'../data/raw/Base_de_Datos_de_Inclusion_Financiera_201903.xlsm', 
                         sheet_name='BD Infraestructura Mun', 
                         header=11,
                         engine='openpyxl')
```

```python
## Clearing the file
df_2019_03=df_2019_03.drop(columns=[col for col in df_2019_03.columns if col.endswith('.1') or col=='Unnamed: 0'])
df_2019_03=df_2019_03.rename(columns={'Clave \nEstado':'Clave_Estado'})
df_2019_03.columns = df_2019_03.columns.str.replace('\n', ' ').str.replace(' ', '_')
df_2019_03['qrt']='2019_03'

print(df_2019_03.head())
print(df_2019_03.shape)
```

```python
## Loading function

def load_base(filepath, quarter):
    df=pd.read_excel(filepath, 
                         sheet_name='BD Infraestructura Mun', 
                         header=11,
                         engine='openpyxl')
    df=df.drop(columns=[col for col in df.columns if col.endswith('.1') or col=='Unnamed: 0'])
    df.columns = df.columns.str.lower()
    df.columns = df.columns.str.replace('\n', ' ').str.replace(' ', '_')
    df=df.rename(columns={'clave__estado':'clave_estado', 'cajeros_automáticos':'cajeros'})
    df['qrt']=quarter
    return df


```

```python
list=[201903, 201906, 201909, 201912,
202003, 202006, 202009, 202012,
202103, 202106, 202109, 202112,
202203, 202206, 202209]

df_list=[]
for x in list:
    df=load_base(f'../data/raw/Base_de_Datos_de_Inclusion_Financiera_{x}.xlsm', f'{x}')
    df_list.append(df)
df=pd.concat(df_list)
print(df.head())
print(df.shape)

```
