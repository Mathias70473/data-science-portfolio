```python

#continuous variable
continuous = ['laufzeit', 'hoehe', 'alter']
continuous_map={'laufzeit':'Duration', 'hoehe':'Amount', 'alter':'Age'}

# categorical
categorical = ['laufkont', 'moral', 'verw', 'sparkont', 'beszeit',
               'rate', 'famges', 'buerge', 'wohnzeit', 'verm',
               'weitkred', 'wohn', 'bishkred', 'beruf', 'pers',
               'telef', 'gastarb']


categorical_map = {
    'laufkont': 'Checking account status',
    'moral': 'Credit history',
    'verw': 'Purpose',
    'sparkont': 'Savings account',
    'beszeit': 'Employment duration',
    'rate': 'Installment rate',
    'famges': 'Personal status_sex',
    'buerge': 'Guarantors',
    'wohnzeit': 'Residence duration',
    'verm': 'Property',
    'weitkred': 'Other installments',
    'wohn': 'Housing',
    'bishkred': 'Existing credits',
    'beruf': 'Job',
    'pers': 'Dependents',
    'telef': 'Telephone',
    'gastarb': 'Foreign worker'
}

target = 'kredit'

target_map = {'kcredit': 'Credit Risk'}
```

```python

for i,b in continuous_map.items():
    plt.figure(figsize=(10,6))
    df[i].hist(bins=30, color='skyblue', alpha=0.7)
    plt.title(f'Variable distribution for {b}', fontsize=14, fontweight='bold')
    plt.xlabel(f'{b}', fontsize=12)
    plt.ylabel('Frequency', fontsize=12)
    plt.grid(axis= 'y', alpha=0.3)
    plt.tight_layout()
    plt.savefig(f'../outputs/figures/{b} distribution.png', dpi=150)
    plt.show()
    
```

```python
for i,b in categorical_map.items():
    plt.figure(figsize=(10,6))
    sns.countplot(data=df, x=i, palette ='Set2')
    plt.title(f'variable {b}', fontsize=14, fontweight='bold')
    plt.xlabel(f'{b}', fontsize=12)
    plt.ylabel('Count', fontsize=12)
    plt.tight_layout()
    plt.savefig(f'../outputs/figures/{b} countplot.png', dpi=150)
    plt.show()
    
    
```

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

for ax, var in zip(axes, continuous):
    df.boxplot(column=var, by='kredit', ax=ax)
    ax.set_title(continuous_map[var])
    ax.set_xlabel('kredit (0=bad, 1=good)')
plt.suptitle('')
plt.tight_layout()
plt.savefig('../outputs/figures/continuous_vs_target.png', dpi=150)
plt.show()
```

```python
# Default rate by category for 3 key categoricals
key_cats = ['laufkont', 'moral', 'sparkont']

fig, axes = plt.subplots(1, 3, figsize=(15, 5))

for ax, var in zip(axes, key_cats):
    default_rate = df.groupby(var)['kredit'].mean()
    default_rate.plot(kind='bar', ax=ax, color="#6395C7")
    ax.set_title(categorical_map[var])
    ax.set_ylabel('Default rate (mean kredit)')
    ax.set_xlabel(categorical_map[var])
    ax.axhline(df['kredit'].mean(), color='red', linestyle='--', label='Overall mean')
    ax.legend()
plt.tight_layout()
plt.savefig('../outputs/figures/categorical_vs_target.png', dpi=150)
plt.show()
```
